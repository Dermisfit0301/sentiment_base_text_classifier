# Text classification with machine learning application

In this assignment, we will create a text classification model that uses NLP techniques to identify sentiments and then sort texts into various categories based on those sentiments.

**Step 1:Preparing the Dataset**

In the first phase we will prepare the dataset for further analysis. We have almost 1 lakh records in this dataset.  We will filter out the empty rows and unnecessary features. We will perform some feature engineering and then we will proceed to the next phase.

```python
import pandas as pd

# Setting the dataset oath
dataset_path = "/content/drive/MyDrive/ASSIGNMENTS/NLP task/amazon_reviews_us_Personal_Care_Appliances_v1_00.tsv"

# Load the dataset into a Pandas DataFrame
adf = pd.read_csv(dataset_path, delimiter='\t', error_bad_lines=False)
adf.head()
```

```python
#examining the dataset
adf.info()
```

```python
#dropping the null values
adf=adf.dropna()
adf
```

```python
#keeping only the necessary columns
adf = adf[['customer_id', 'review_body', 'star_rating']]
adf.head()
```

```python
#seeing the number of star ratings
adf['star_rating'].value_counts()
```

```python
# Making a function to map star ratings to sentiment labels
def map_star_to_sentiment(star_rating):
    if star_rating == 5:
        return 'positive'
    elif star_rating == 4:
        return 'positive'
    elif star_rating == 3:
        return 'neutral'
    elif star_rating == 2:
        return 'negative'
    elif star_rating == 1:
        return 'negative'
    else:
        return 'unknown'
```

```python
# Applying the mapping function and making a new 'sentiment' column
adf['sentiment'] = adf['star_rating'].apply(map_star_to_sentiment)
adf['sentiment'].head()
```

```python
#dropping the unnecessary column
adf=adf.drop(['star_rating'],axis=1)
adf.head()
```

```python
#seeing which sentiment is popular
adf['sentiment'].value_counts()
```

**Step 2: NLP phase**

In this step we are going to use NLP techniques to pre process the data so that it can be used for effective labelling. Activities like text cleaning, tokenization, and various transformation would be performed.


```python
#getting the necessary NLP packages
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Making a function for text preprocessing
def preprocess_text(text):

    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'[^a-zA-Z]', ' ', text)  # Keep only letters


    words = word_tokenize(text)


    stop_words = set(stopwords.words('english'))
    filtered_words = [word.lower() for word in words if word.lower() not in stop_words]


    cleaned_text = ' '.join(filtered_words)
    return cleaned_text
```

```python
# Applying the function to the reviews
adf['cleaned_review'] = adf['review_body'].apply(preprocess_text)
adf['cleaned_review'].head()
```

```python
#dropping the useless variable

adf=adf.drop(['review_body'],axis=1)
adf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Getting the TF-IDF vectorizer to activate
tfidf_vectorizer = TfidfVectorizer(max_features=10000)

# Fitting and transforming the cleaned text data using TF-IDF
tfidf_features = tfidf_vectorizer.fit_transform(adf['cleaned_review'])
```

**Step 3: Building the Classifier Model**

In this phase, a sentiment classifier will be designed to meticulously categorize reviews based on their underlying sentiments.

```python
from sklearn.model_selection import train_test_split

# Splitting the data into features and labels
X = tfidf_features
y = adf['sentiment']

```

```python
# Splitting the data into training and testing  sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#getting the Machine learning packages
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

```

```python
#Training the conveentional machine learning models and seeing their accuracy
decision_tree_classifier = DecisionTreeClassifier(random_state=42)
decision_tree_classifier.fit(X_train, y_train)


naive_bayes_classifier = MultinomialNB()
naive_bayes_classifier.fit(X_train, y_train)




# Predicting with the trained models
decision_tree_predictions = decision_tree_classifier.predict(X_test)
naive_bayes_predictions = naive_bayes_classifier.predict(X_test)


# Evaluation of the models
decision_tree_accuracy = accuracy_score(y_test, decision_tree_predictions)
naive_bayes_accuracy = accuracy_score(y_test, naive_bayes_predictions)


print("Decision Tree Accuracy:", decision_tree_accuracy)
print("Naive Bayes Accuracy:", naive_bayes_accuracy)
```

*The conventional machine learning models have a good accuracy but it is good to see if we can make it better so we will use the neural network model.*

```python
from sklearn.neural_network import MLPClassifier

# Initializing the ANN classifier
classifier = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000)
```

```python

# Training the classifier on the training data
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Getting predictions on the testing data
y_pred = classifier.predict(X_test)

# Evaluation of the model's performance
accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)
confusion_mat = confusion_matrix(y_test, y_pred)

# Getting the accuracy, classification report, and confusion matrix
print(f'Accuracy: {accuracy:.2f}')
print(classification_rep)
print('Confusion Matrix:\n', confusion_mat)
```

*The classifier has a good accuracy but it is good to see if we can make it better.It has an accuracy of 81% and does good at predicting the positive class however it doesnt predict the neutral values well. This is because the dataset is imbalanced.We definitely need to perform hyperparameter tuning for better accuracy!*

```python


# Defining the hyperparameters
param_grid = {
  'hidden_layer_sizes': [50, 100],
  'activation': ['relu', 'tanh'],
  'learning_rate': ['constant', 'adaptive'],
  'max_iter': [100, 200],
}

```

```python
# Limiting the dataset to the first 1000 rows due to constraints
X_train_subset = X_train[:1000]
y_train_subset = y_train[:1000]
```

```python
# Define a shared variable to indicate if the timeout is reached
timeout_reached = False

# Function to run random search
def run_random_search():
    global timeout_reached
    try:
        grid_search.fit(X_train_subset, y_train_subset)
    except TimeoutError:
        timeout_reached = True

# Create a thread to run random search
search_thread = threading.Thread(target=run_random_search)

# Starting the thread
search_thread.start()
search_thread.join(timeout=180)

# Checking if the thread is still running
if search_thread.is_alive():

    timeout_reached = True

```

```python


# Get the best hyperparameters

try:
    best_params = grid_search.cv_results_['params'][grid_search.cv_results_['mean_test_score'].argmax()]
except AttributeError:
    print('The GridSearchCV object has not been fitted yet.')


```

```python
#getting the necessary packages for the tuning
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

```

```python
# Creating a Random Forest classifier
rf_classifier = RandomForestClassifier(random_state=42)

# Defining the hyperparameter grid
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Initializing Grid Search with cross-validation


grid_search = GridSearchCV(rf_classifier, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
```

```python
# Fitting the model on 10000 rows

grid_search.fit(X_train[:10000], y_train[:10000])
```

```python
# Getting the best hyperparameters
best_params = grid_search.best_params_
print("Best Hyperparameters:", best_params)
```

```python
# Creating a Random Forest classifier with the best hyperparameters
best_rf_classifier = RandomForestClassifier(
    n_estimators=100,
    max_depth=30,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)

# Fitting the model on the training data
best_rf_classifier.fit(X_train[:10000], y_train[:10000])
```

```python
# Making predictions on the test data
y_pred = best_rf_classifier.predict(X_test[:10000])

# Calculating the metrics
accuracy = accuracy_score(y_test[:10000], y_pred[:10000])
print(f"Accuracy: {accuracy:.2f}")
```
